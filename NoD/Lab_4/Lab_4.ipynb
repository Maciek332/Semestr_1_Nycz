```python
import matplotlib . pyplot as plt
```

```python
#Wykres liniowy

x = [1 , 2 , 3 , 4 , 5]
y = [2 , 3 , 5 , 7 , 11]


plt.plot(x , y , marker = 'o' , linestyle = '-' , color = 'b' , label = 'Liniowy' )
plt.title(" Wykres Liniowy")
plt.legend()
plt.show()

```

```python
#Wykres słupkowy

kategorie = ['A', 'B', 'C', 'D']
wartosci = [5,2,7,1]

#Tworzenie wykresu słupkowego

plt.bar(kategorie, wartosci, color="blue")
plt.xlabel("Kategorie")
plt.ylabel("Wartości")
plt.title("Wykres słupkowy")
plt.show()
```

```python
#histogram
import numpy as np

#Dane

dane = np.random.normal(0, 1, 1000)

#tworzenie histogramu

plt.hist(dane, bins=30, color="green", alpha=0.7)
plt.xlabel("Wartości")
plt.ylabel("Częstotliwość")
plt.title("Histogram")
plt.show()
```

```python
#wykres kołowy
kategorie = ['A', 'B', 'C', 'D']
wartosci = [15,22,74,12]

#tworzenie wykresu kołowego

plt.pie(wartosci, labels=kategorie, autopct="%1.1f%%", startangle=90)
plt.title("Wykres kołowy")
plt.show()
```

```python
#Dostosowywanie wykresów

#kolory, linie i style

x = [1,2,3,4,5]
y=[10, 20, 25, 30, 35]

plt.plot(x,y, color="orange", marker="x", linestyle="--", linewidth=2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Dostosowany wykres liniowy")
plt.show()
```

```python
#Dodawanie siatki

plt.plot(x,y,color="purple", marker="o")
plt.grid(True)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres z siatką")
plt.show()
```

```python
#dodawanie adnotacji

plt.plot(x,y,marker="o")
plt.annotate("Najwyższy punkt", xy=(5,35), xytext=(3, 30), arrowprops=dict(facecolor="pink", arrowstyle="->"))
plt.title("Wykres z adnotacją")
plt.show()
```

```python
#Wykres punktowy

x = np.random.rand(50)
y = np.random.rand(50)

sizes = 1000 * np.random.rand(50)

colors = np.random.rand(50)

plt.scatter(x,y,s=sizes, c=colors, alpha=0.5, cmap="viridis")
plt.colorbar()
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Wykres punktowy")
plt.show()
```

```python
#wykres 3d

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y ** 2))

ax.plot_surface(x, y, z, cmap='plasma')
plt.title("Wykres 3D")
plt.show()
```

```python
#wykres kołowy plotly
import plotly.express as px
#dane
x = [1,2,3,4,5]
y=[10,15,24,13,23]
#tworzeniw wykresu
fig = px.line(x=x, y=y, title="wkres kołowy", labels={'x':"0 x", 'y':"0 y"})
fig.show()
```

```python
#wykres słupkowy
kategorie = ['A', 'B', 'C', 'D']
wartosci = [4,5,7,8]

#Tworzenie wykresu słupkowego

fig = px.bar(x=kategorie, y=wartosci, title="Wykres słupkowy", labels={'x': 'Kategorie', 'y': "Wartości"})
fig.show()
```

```python
#Wykres kołowy

fig = px.pie(names=kategorie, values=wartosci, title="Wykres kołowy")
fig.show()
```

```python
#Histogram

import numpy as np

dane = np.random.normal(0, 1, 1000)

#tworzenie hostogramu
fig = px.histogram(x=dane, nbins=30, title="Histogram", labels={'x': 'Wartości', 'y': "Częstotliwość"})
fig.show()
```
